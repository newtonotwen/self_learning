```python
import matplotlib.pyplot as plt
import numpy as np
```

<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 200)
tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
# code here
fig, (ax1,ax2, ax3) = plt.subplots(1,3, figsize=(15,6), sharey= True)

ax1.plot(x, tan, color='b', label='Tanx')
ax1.set_title('Garfik Tanx')
ax1.legend()

ax2.plot(x, cos, color='r', label='Sinx')
ax2.set_title('Grafik Sinx')
ax2.legend()

ax3.plot(x, sin, color='g', label='Cosx')
ax3.set_title('Grafik Cosx')
ax3.legend()


plt.tight_layout()
plt.show()
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
# code here
fig, ax= plt.subplots(figsize= (14,10))

ax.plot(x,y)
ax.set_title('Grafik Cosx')

ins1=ax.inset_axes([0.9, 0.9, 0.3, 0.3])
ins1.plot(x, np.cos(x))
ins1.plot(x, y4, color='k', linewidth=1)
ins1.set_title('Grafik Cosx^4')

ins2=ax.inset_axes([-0.1, 0.9, 0.3, 0.3])
ins2.plot(x, np.cos(x))
ins2.plot(x, y5, color='y', linewidth=1)
ins2.set_title('Grafik Cosx^5')

ins3 = ax.inset_axes([-0.1, -0.1, 0.3, 0.3])
ins3.plot(x, np.cos(x))
ins3.plot(x, y2, color='r', linewidth=1)
ins3.set_title('Grafik Cosx^2')

ins4 = ax.inset_axes([0.9, -0.1, 0.3, 0.3])
ins4.plot(x, np.cos(x))
ins4.plot(x, y3, color='g', linewidth=1)
ins4.set_title('Grafik Cosx^3')

plt.tight_layout()
plt.show()
```
